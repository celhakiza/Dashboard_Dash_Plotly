```python
## start an app
from dash import Dash
from dash import dcc
from dash import html
import plotly.express as px
from jupyter_dash import JupyterDash
from dash.dependencies import Input, Output
import pandas as pd
import dash_bootstrap_components as dbc
```

```python
data=pd.read_clipboard()
```

```python
app=JupyterDash(__name__)
app.layout=html.Div(
                [
    
                html.H1("My first App"),
                html.H2("Superstore Sales and Profits"),
                dcc.Markdown('''
                            this is the dataset provided by USA, we use it for study purpose
                            and you get insight for the us real data that you can use to enhance your skills
                            practice it more and more you get better results at the end. the very first graph shows 
                            the sales and profits across region of USA. the next graph shows the sales of categories 
                            (office supplies, technology and furniture) in different region. the last graph shows the sales 
                            across state of USA on a map
                                
                             '''
                            ),
                    
                                
                        dcc.Dropdown(
                            options=["Sales","Profits"], 
                            value="Sales",
                            id="sales-profits-dropdown"),
                            html.Div(id='sales-profits-display'),
                            
                        html.H2("Sales across region"),
                            
                        dcc.Dropdown(
                            options=["Central","East","South","West"],
                            value="South",
                            id="Region-dropdown"),
                        html.H1("Profits by State"),
                        html.Div(id="region-sales-display"),
                        dcc.RangeSlider(
                                       min=2014,
                                       max=2018,
                                       value=[2016,2017],id='slider-year',
                                       marks={str(c):str(c) for c in range(2014,2018,1)}),

                        dcc.Dropdown(
                            options=["Furniture","Office Supplies","Technology"],
                            value="Furniture",
                            id="category-dropdown"),

                            html.Div(id="category-display")

                               

                                
@app.callback(
            Output("sales-profits-display","children"),
            Input("sales-profits-dropdown","value")
)
def update_graph(select_cat):
    dataa1=data.groupby(['Region','Category'],as_index=False)['Sales','Profit'].sum()
    if select_cat=="Sales":
        fig=px.bar(dataa1,x='Region',y='Sales',color='Category',barmode="group",title="Sales across Region",height=500,width=700)
    elif select_cat=="Profits":
        fig=px.bar(dataa1,x='Region',y='Profit',color='Category',barmode="group",title="Profits across Region",height=500,width=700)
      
    return dcc.Graph(figure=fig)
@app.callback(
        Output("region-sales-display","children"),
        Input("Region-dropdown","value")

)
def bar_cat(cat):
    dataa=data.groupby(['year_ship','Category','Region'],as_index=False)['Sales','Profit','Quantity'].sum()
    if cat=="Central":
        fig=px.bar(data_frame=dataa[dataa['Region']=='Central'], x='year_ship',y='Sales',height=500,width=700,
            barmode='group',
            color='Category')
    elif cat=="East":
        fig=px.bar(data_frame=dataa[dataa['Region']=='East'], x='year_ship',y='Sales',height=500,width=700,
            barmode='group',
            color='Category')
    elif cat=="South":
        fig=px.bar(data_frame=dataa[dataa['Region']=='South'], x='year_ship',y='Sales',height=500,width=700,
            barmode='group',
            color='Category')
    elif cat=="West":
        fig=px.bar(data_frame=dataa[dataa['Region']=='West'], x='year_ship',y='Sales',height=500,width=700,
            barmode='group',
            color='Category')
        
    return dcc.Graph(figure=fig)

@app.callback(
            Output("category-display","children"),
            Input("category-dropdown","value")
)
def visualize_map(state):
    if state=="Furniture":
        data5=data[data['Category']=='Furniture']
        fig =px.choropleth(
            data_frame=data5,
            locations='State_code',
            locationmode='USA-states',
            color='Sales',
            projection='albers usa',
            color_continuous_scale=px.colors.sequential.Oranges)
    elif state=="Office Supplies":
        data6=data[data['Category']=='Office Supplies']
        fig =px.choropleth(
            data_frame=data6,
            locations='State_code',
            locationmode='USA-states',
            color='Sales',
            projection='albers usa',
            color_continuous_scale=px.colors.sequential.Oranges)
    
    elif state=="Technology":
        data7=data[data['Category']=='Technology']
        fig =px.choropleth(
            data_frame=data7,
            locations='State_code',
            locationmode='USA-states',
            color='Sales',
            projection='albers usa',
            color_continuous_scale=px.colors.sequential.Oranges)
        
    return dcc.Graph(figure=fig)
        
    
app.run_server()  
```
